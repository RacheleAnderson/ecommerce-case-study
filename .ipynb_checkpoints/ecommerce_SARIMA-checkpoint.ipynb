# Data Science for E-Commerce
## Sales predictions
This notebook is dedicated to the sales predictions using Seasonal ARIMA models (SARIMA).

Refer to ecommerce_explanatory.ipynb for the exploratory analysis, and to ecommerce_insights.ipynb for data insights.

#### Data cleaning and preprocessing
Clean data and split it into training, validation and test data. Assuming a yearly cycle, we use as training + validation the first 365 days (training 300 days, validation 65 days) and for test data the first days of december 2011.

```python
# import libraries
#import importlib
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.api import qqplot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from pmdarima.preprocessing import BoxCoxEndogTransformer
import pmdarima as pm
import preprocess_data

#importlib.reload(preprocess_data)
```

```python
data = pd.read_csv('ecommerce_data.csv')

# preprocess data
data, data_orig_no_dup, train_and_validation_data, test_data = preprocess_data.preprocess_data(data)
```

### Time series analysis
A time series analysis model uses historical data to forecast the future. 

ARIMA, short for ‘Auto Regressive Integrated Moving Average’ is a class of models that models a given time series based on its own past values, so that equation can be used to forecast future values.
An ARIMA model is characterized by 3 terms: p, d, q, where:
- p is the order of the AR term
- q is the order of the MA term
- d is the number of differencing required to make the time series stationary

If there is a seasonal pattern we add seasonal terms and it becomes SARIMA, short for ‘Seasonal ARIMA’. 

### Step 1 - Check if data is stationary
The data need to be stationary for proceed with ARIMA models. First we check stationarity and decide if a transformation is needed.
Transformation of the data can be needed if the variance of the data changes over time.
Differentiation can be needed if there are strong trends in the data.

#### Check if transformation is needed.
Plot data, ACF & histogram.

```python
pm.tsdisplay(train_and_validation_data.values, lag_max=50, title="Sales Data", show=True)
```

The histogram is skewed, should preferably look similar to a gaussian distribution.
Let's transform it using BoxCox transformation.

```python
# boxcox can't have values <= 0,  lmbda2 is an constant offset added to the data to remove any 0 values.
boxcox_model = BoxCoxEndogTransformer(lmbda2=1e-6)
# train_val_bc is the boxcox transformed train_and_validation_data
train_val_bc, _ = boxcox_model.fit_transform(train_and_validation_data)
print(f"Best BoxCox Lambda fit: {boxcox_model.lam1_}")

pm.tsdisplay(
    train_val_bc, lag_max=50, title="Sales Data (BoxCox-transformed)", show=True)
```

This looks better.
The histogram resembles one from data gaussian distributed.
The spike at -5 is data that was 0 before (saturdays and holidays). It is not a problem for continuing modeling.

#### Test of stationarity
Let's check if the data is stationary after the BoxCox transformation.
Here we test stationarity of:
1. Unmodified data
2. BoxCox transformed data
3. Differentiated data

```python
output=adfuller(train_and_validation_data)
print('Stat check on original data')
print(f"ADF Stats: {output[0]}, P-value: {output[1]}")
print("p-values > 0.05, can't reject null hypothesis -> assume not stationary")

print("--"*30)
output=adfuller(train_val_bc)
print('Stat check on BC data')
print(f"ADF Stats: {output[0]}, P-value: {output[1]}")
print("p-values < 0.05, reject null hypothesis -> assume stationary")

print("--"*30)
diff_train_val = train_and_validation_data.diff(1)
diff_train_val = diff_train_val.dropna()
output = adfuller(diff_train_val)
print('Stat check on diff data')
print(f"ADF Stats: {output[0]}, P-value: {output[1]}")
print("p-values < 0.05, reject null hypothesis -> assume stationary")
```

As expected the unmodified data is not stationary.
Since BC-transformed data is stationary, it might not be necessary to differentiate in this step.

### Step 2 - Look at the ACF and PACF transformed data
To gain insight of the data sesonality we can look at the ACF and PACF of the data.
For comparison we plot the ACF and PACF for both unmodified and transformed data.

```python
def plot_acf_pacf(y, lags=None, suptitle=None, figsize=None):
    """
    Helper function that plots the acf and pacf
    :param y: Data
    :param lags: optional, number of lags to calculate and plot (ACF/PACF)
    :param suptitle: title of plot
    :param figsize: See matplotlib documentation
    :return:
    """
    fig, ax = plt.subplots(1,2,figsize=figsize)
    plot_acf(y, lags=lags, ax=ax[0])
    plot_pacf(y,lags=lags, ax=ax[1])
    fig.suptitle(suptitle)
    plt.show()

# ACF/PACF transformed data
plot_acf_pacf(train_val_bc, figsize=(12,4), suptitle='ACF and PACF of the BoxCox transformed data')
# ACF/PACF unmodified data
plot_acf_pacf(train_and_validation_data, figsize=(12,4), suptitle='ACF and PACF of the unmodified data')
```

There is an obvious strong seasonality at lag 7.
Overall there are not much difference between the ACF/PACF of the transformed and unmodified data.
In some cases this could indicate that we could have gotten away with modelling on the unmodified data.
I tried it, this is not the case. So we will continue with the transformed data.

### Step 3 - Split training and validation data for model development
For the stationary check, the train and validation data has been one dataset.
Time to split them apart. We use the first 300 data points for training and the last 65 for validation.

```python
# Number of samples in the training set
n_train_samples = 300
# Redefine the boxcox transformer
boxcox_model = BoxCoxEndogTransformer(lmbda2=1e-6)

# Split the train and validation data
train = train_and_validation_data[:n_train_samples]
validation = train_and_validation_data[n_train_samples:]
boxcox_model.fit(train)

# "_bc" denotes BoxCox transformed data
train_bc, _ = boxcox_model.transform(train)
validation_bc, _ = boxcox_model.transform(validation)

# To match data later we define some variables containing the index of the data in the train and validation data in relation to train_and_validation_data
# Integer index of train_and_validation_data
idx_complete = np.arange(len(train_and_validation_data))
# Index of data in train
idx_train = idx_complete[:n_train_samples]
# Index of data in validation
idx_validation = idx_complete[n_train_samples:]

# Datetime index
days = list(train_and_validation_data.index)
days_train = days[:n_train_samples]
days_test = days[n_train_samples:]
```

### Step 4 - SARIMA Model of the time series
We will now model the data using a Seasonal Autoregressive Integrated Moving Average (SARIMA) model.
From the ACF and PACF in "step 2" it's evident that the model should contain a seasonal part of 7.

The rest of the model order parameters is an iterative process.
To simplify it I first created a model using the autoSARIMA function in the "pmdarima" package.
Using the result from that as a base I tried to improve the model by tuning the orders.
(which in the end was unecessary, as no significantly better model could be found).

Below the final model is implemented using the statsmodels package.

More information on the stat SARIMAX function & state space model can be found here:
https://www.statsmodels.org/devel/generated/statsmodels.tsa.statespace.sarimax.SARIMAX.html

```python
model = SARIMAX(endog=train_bc, order=(1,1,1), seasonal_order=(1, 0, 1, 7), trend=None)
model_fit = model.fit(maxiter=100)
print(model_fit.summary())
```

#### Model diagnostics

```python
model_fit.plot_diagnostics(figsize=(12,8), lags=40)
plt.show()
```

It looks pretty good.
The histogram shows that the residuals looks almost normally distributed.
The Normal Q-Q plot has something going on in the tails.
The upper tail are most likely due to the fact that we have not removed outliers.
The lower tail is very likely due to all the zeros in the data set.
Especially zeros happening in the middle of the week due to holidays.

The correlogram indicates that there is no more structure to be modelled in the data.

### Step 5 - Sanity check predictions
Now we will look at how the SARIMA model handles predictions.

```python
# .predict(): predictions of period t values conditional on observed data through period t-1. More info in the link below.
# https://www.statsmodels.org/devel/generated/statsmodels.tsa.statespace.sarimax.SARIMAXResults.predict.html?highlight=predict#statsmodels.tsa.statespace.sarimax.SARIMAXResults.predict
# model_fit.predict() -> returns 1-step predictions on the data given to the model as training data.
# i.e. train_pred[5] is the 1-step prediction of t=5 using data up to t=4.
train_pred = model_fit.predict()

# .forecast(): perform one-step-ahead forecast, i.e. creates a one step prediction based on all the data observed by the model.
# if we want to do 1-step predictions in a similar fashion on the validation set we will have to add the observations to the model.
# this can be done using model_fit.append(data_point). Good example of this can be found here:
# https://www.statsmodels.org/devel/examples/notebooks/generated/statespace_forecasting.html

model_fit_copy = model_fit # Shallow copy, needed to be able to re-run the cell.
# Store validation predictions
validation_pred = np.array([])

for value in validation_bc:
    validation_pred = np.append(validation_pred,model_fit_copy.forecast())
    model_fit_copy = model_fit_copy.append([value], refit=False) # [value] because .append here needs a object with a length

#print(model_fit.nobs)
def bc_inverse_transform(y, bc_model):
    """
    Helper function for inverse transform boxcox.
    This is needed because numpy does not allow fractional powers of negative numbers.
    More or less copy-paste from the inverse_transform function given in the package.

    :param data: Data to be inverse transformed
    :param bc_model: fitted boxcox model
    :return: Inverse transformed data
    """
    lam1 = bc_model.lam1_
    lam2 = bc_model.lam2_


    numer = y*lam1 # remove denominator
    numer += 1 # add 1 back to it
    sign_numer = np.sign(numer)
    de_exp = np.abs(numer) ** (1. / lam1) # de-exponentiate
    de_exp = sign_numer*de_exp
    return de_exp - lam2

# Note predictions is in the boxcox transformed "domain", inverse transform to compare with unmodified data
train_pred_inverse = bc_inverse_transform(train_pred, boxcox_model)
validation_pred_inverse = bc_inverse_transform(validation_pred, boxcox_model)

# Plot the predictions vs actual data

fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(idx_train, train, label='Train Data', lw=2)
ax.plot(idx_train, train_pred_inverse, label='Train - In sample prediction', ls='--')
ax.plot(idx_validation, validation, label="Validation Data", lw=2)
ax.plot(idx_validation, validation_pred_inverse, label="Validation - predictions", ls='--', c='red')
ax.legend()
ax.set_title('Comparison predicted values and actual data')
#plt.savefig('model_pred_val.png', dpi=300, bbox_inches="tight")
plt.show()

# Calculate Mean square errors of the residuals
train_residuals =train - train_pred_inverse
validation_residuals = validation - validation_pred_inverse
print('MSE Train :',np.sqrt(np.mean((train_residuals)**2)))
print('MSE Validation :',np.sqrt(np.mean((validation_residuals)**2)))

print("--"*40)
# Lets also plot the ACF/PACF and QQ plot of the prediction residuals
plot_acf_pacf(train_residuals, suptitle="ACF & PACF Train Residuals", figsize=(12,4))
plot_acf_pacf(validation_residuals, suptitle="ACF & PACF Validation Residuals", figsize=(12,4))

fig, axes = plt.subplots(1, 2, figsize=(12,4))

qqplot(train_residuals, line='s', fit=True,ax=axes[0])
axes[0].set_title("Normal Q-Q - Train residuals")
qqplot(validation_residuals, line='s', fit=True,ax=axes[1])
axes[1].set_title("Normal Q-Q - Train residuals")

plt.show()
```

From the plots above we can say that the model works pretty well in performing one-step predictions.
One could possible want better looking Q-Q plots but unfortunately this is often the reality when working with real data.

### Step 6 - Forecast the testset
We will now re-train the model using all the data in train/validation set, then forecast the data in the test set (december 2011).

```python
# Lets put everything here needed to train the model.

boxcox_model = BoxCoxEndogTransformer(lmbda2=1e-6)
train_val_bc, _ = boxcox_model.fit_transform(train_and_validation_data)

# The model
model = SARIMAX(endog=train_val_bc, order=(1,1,1), seasonal_order=(1, 0, 1, 7), trend=None)
model_fit = model.fit(maxiter=100)
print(model_fit.summary())
# final model diagnostics
model_fit.plot_diagnostics(figsize=(12,8)) 
#plt.savefig('model_diagnostics.png', dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Same train test plot as before
n_test_samples = len(test_data)

# get_forecast(n_test_samples) will forecast up to n_test_samples days in the future.
# it also returns confidence intervals
test_forecast_frame = model_fit.get_forecast(n_test_samples).summary_frame()
test_forecast = test_forecast_frame['mean']
conf_int_lower = test_forecast_frame['mean_ci_lower']
conf_int_upper = test_forecast_frame['mean_ci_upper']

# Again want to transform from boxcox "domain"
test_forecast_inverse = bc_inverse_transform(test_forecast, boxcox_model)
conf_int_lower_inverse = bc_inverse_transform(conf_int_lower, boxcox_model)
conf_int_upper_inverse = bc_inverse_transform(conf_int_upper, boxcox_model)

# Lets look at the predictions
fig, ax = plt.subplots(figsize=(12, 4))
# Define integer base index for the test set.
idx_test = idx_complete[-1]+1+np.arange(n_test_samples)
# Plot last 50 samples of train/val data (for cleaner plot), and all test data.
ax.plot(idx_complete[-50:], train_and_validation_data[-50:], label='Tail of Train/Validation Data')
ax.plot(idx_test, test_data, label="Test data")
ax.plot(idx_test, test_forecast_inverse, label="Test forecast", ls='--', lw=3)
ax.fill_between(idx_test, conf_int_lower_inverse, conf_int_upper_inverse, alpha=0.1)
fig.suptitle('Prediction on test data')
plt.legend(loc='upper left')
#plt.savefig('model_pred_test.png', dpi=300, bbox_inches="tight")
plt.show()

# Compare to predictions before transformation.
fig, ax = plt.subplots(figsize=(12, 4))
# Plot last 50 samples of train/val data (for cleaner plot), and all test data.
ax.plot(idx_complete[-50:], train_val_bc[-50:], label='Tail of Train/Validation Data')
ax.plot(idx_test, boxcox_model.transform(test_data)[0], label="Test data")
ax.plot(idx_test, test_forecast, label="Test forecast", ls='--', lw=3)
ax.fill_between(idx_test, conf_int_lower, conf_int_upper, alpha=0.1)
fig.suptitle('Prediction on test data in BoxCox domain')
plt.legend(loc='upper left')
#plt.savefig('model_pred_test_BC.png', dpi=300, bbox_inches="tight")
plt.show()
```

#### Comments on this model and predictions
The most noteworthy detail in the plots above is the fact that the confidence interval explodes after inverse transforming the original domain.
This is not something to care too much about, as the actual predictions are pretty good.
If we look at the predictions in the boxcox "domain", the confidence interval is more reasonable.

Since the test set is so small there is not much more that can be done as ACF/PACF etc would need a bit more data to create them.

To summarize, it is possible to use a SARIMA model to forecast sales.
However, it would need more fine-tuning, and some "intelligence".
One trivial thing one should add is a "if-statement" that sets forecast to 0 whenever there are holidays.

One could also use the dates of the holidays as a known exogenous signal, and add it to the model.
Then it might be possible for the model to learn that it should be zero at these times.
